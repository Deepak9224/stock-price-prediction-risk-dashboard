```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
stocks = {
    "Reliance": "RELIANCE.NS",
    "Infosys": "INFY.NS",
    "HDFC_Bank": "HDFCBANK.NS",
    "ITC": "ITC.NS",
    "Maruti": "MARUTI.NS",
    "SBI": "SBIN.NS"
}

benchmark = "^NSEI"   
```

```python
for company, ticker in stocks.items():

    df = yf.download(
        ticker,
        start="2019-01-01",
        end="2025-12-31"
    )

    df.to_csv(f"../data/raw/{company}.csv")

    print(f"{company} data saved")
```

```python
nifty = yf.download(
    benchmark,
    start="2019-01-01",
    end="2025-12-31"
)

nifty.to_csv("../data/raw/NIFTY50.csv")

print("NIFTY data saved")
```

Creating Empty Master DataFrame

```python
master_df=pd.DataFrame()
```

```python
for company in stocks.keys():

    df = pd.read_csv(
        f"../data/raw/{company}.csv",
        skiprows=2
    )

    df.columns = [
        "Date",
        "Close",
        "High",
        "Low",
        "Open",
        "Volume"
    ]

    df['Company'] = company

    master_df = pd.concat([master_df, df])
```

```python
master_df.head()
```

```python
master_df.shape
```

```python
master_df.to_csv(
    "../data/processed/master_stock_data.csv",
    index=False
)
```
